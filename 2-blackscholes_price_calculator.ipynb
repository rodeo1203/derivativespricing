```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from scipy.stats import norm
import yfinance as yf 
import pandas_datareader as web
from datetime import datetime as dt, timedelta as td
import seaborn as sns
sns.set()
```

```python
def bsm(K, stock='tsla', r=.015, T = 1):
    df = yf.download(stock, start=dt.today() - td(days=365*10))['Adj Close']
    global prii
    prii=df
    log_ret = np.log(df.pct_change() + 1)
    std = log_ret.std()*250**.5
    S = df.iloc[-1]
    d1 = (np.log(S/K) + (r + (std ** 2 / 2))*T) / (std * np.sqrt(T))
    d2 = d1 - (std * np.sqrt(T))
    call = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    put = - (S*norm.cdf(-d1) - K*np.exp(-r*T)*norm.cdf(-d2))
    return call, put

    
```

```python
bsm(210)
```

```python
for i in range(1, 10):
    print(f"No. of years {i}, have (call, put) price {bsm(210,'tsla',.015,i)}")
```

Larger the period, more the chance of volatility and hence it is expected that call option price would rise. We shall observe this below.
